# Plotting

```python
import pandas as pd
# initialize data set and preview format
ramenData = pd.read_csv('ramen-ratings.csv')
# change index to review number and sort the data frame
ramenData.set_index('Review #', inplace=True)
ramenData.sort_index(inplace=True)
ramenData.head()
```

```python
import matplotlib as plt
# view plots in the notebook
%matplotlib inline
```

## Plot the Total Ramen Varieties by Country

```python
ramenPerCountry = ramenData[['Variety', 'Country']]
# create a monochromatic bar graph, sized appropriately
ramenPerCountry.Country.value_counts().plot(kind= 'bar', figsize= (10,5), colormap = 'Paired');
```

## Make a Pie Chart Showing Rating Distribution for US Ramen

```python
usRamenRatings = ramenData[['Country', 'Stars']]
usRamenRatings = ramenData[(ramenData.Country == 'USA')]
usRamenRatings.Stars.value_counts().plot(kind= 'Pie', figsize= (15,15), colormap= 'YlGn_r');
```
