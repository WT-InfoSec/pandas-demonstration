# Indexing

Indexing in pandas allows you to access a row or column using a label. 
The index object is an immutable array, and does not support mutable operations.

```python
import pandas as pd
# initialize data set and preview format
ramenData = pd.read_csv('ramen-ratings.csv')
ramenData.head()
```

## Set the Index of the Data Frame to the Review Number

```python
reviewIndex = ramenData.set_index('Review #')
reviewIndex.head()
```

## Sort the Review Index so that #1 is at the Top

```python
reviewIndex.sort_index(inplace= True)
reviewIndex
```

## Retrieve Ramen Reviews from #600 to #620

```python
reviewIndex.loc[600:620]
```

## Retrieve Ramen Reviews from a List of Review Numbers

```python
desiredReviews = [1,42,560,783,874,1240,2097,2428]
reviewIndex.loc[desiredReviews]
```
