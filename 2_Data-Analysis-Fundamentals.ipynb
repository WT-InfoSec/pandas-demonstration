# Data Analysis Fundamentals

```python
import pandas as pd
# initialize data set and preview format
ramenData = pd.read_csv('ramen-ratings.csv')
# change index to review number and sort the data frame
ramenData.set_index('Review #', inplace=True)
ramenData.sort_index(inplace=True)
ramenData.head()
```

## Create Series Displaying Number of Ramen Varieties Per Brand

### Display Brands from Greatest to Least Varieties of Ramen

```python
ramenData['Brand'].value_counts()
```

### Display Brands from Least to Greatest Varieties of Ramen

```python
ramenData.Brand.value_counts(ascending=True)
```

# Sorting

## Create Data Frame of Ramen Brands and Varieties (sorted alphabetically)

```python
ramenVarieties = ramenData[['Brand','Variety']]
ramenVarieties.sort_values(by= ['Brand', 'Variety'])
```

## Create Data Frame Sorted by Ramen Brand and Rating

```python
ratingsPerBrand = ramenData.sort_values(by=['Brand','Stars'])
ratingsPerBrand
```

## Create Data Frame With All 5-Star Ramen Varieties

```python
fiveStarRamen = ramenData[ramenData.Stars == '5']
fiveStarRamen
```

## Sort 5-Star Data Frame by Country of Origin, and Alphabetically by Brand

```python
fiveStarRamen.sort_values(by= ['Country', 'Brand'])
```

## Create Data Frame With Ramen From the Top Ten List (using boolean operators)

```python
# eliminate null values and new line string literals in the "Top Ten" column
topTenRamen = ramenData[(ramenData['Top Ten'].notnull()) 
                        & (ramenData['Top Ten'] != '\n')]
# sort data frame chronologically
topTenRamen.sort_values(by= 'Top Ten')
```

# String Handling

## Create a Data Frame With All Shrimp Flavored Ramen

```python
shrimpRamenSeries = ramenData[ramenData.Variety.str.contains('Shrimp')]
shrimpRamenSeries
```

## Create a Data Frame with All 5-Star, Shrimp Flavored Ramen From Thailand

```python
bestThaiShrimp = ramenData[(ramenData.Stars == '5')
                            & (ramenData.Variety.str.contains('Shrimp')) 
                            & (ramenData.Country.str.startswith('Thai'))]
bestThaiShrimp
```
